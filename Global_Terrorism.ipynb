```python
year_count = df["iyear"].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.plot(year_count.index, year_count.values)
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title("Year-wise Terrorist Attacks")
plt.show()
```

```python
region_count = df["region_txt"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(region_count.index, region_count.values)
plt.xticks(rotation=90)
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.title("Region-wise Terrorist Attacks")
plt.show()

```

```python
country_count = df["country_txt"].value_counts().head(15)

plt.figure(figsize=(8,5))
plt.bar(country_count.index, country_count.values)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Number of Attacks")
plt.title("Top 10 Most Affected Countries")
plt.show()
```

```python
attack_type = df["attacktype1_txt"].value_counts()

plt.figure(figsize=(6,10))
plt.pie(attack_type.values, labels=attack_type.index, autopct="%1.1f%%")
plt.title("Attack Type Distribution")
plt.show()
```

```python
weapon_count = df["weaptype1_txt"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(weapon_count.index, weapon_count.values)
plt.xticks(rotation=90)
plt.xlabel("Weapon Type")
plt.ylabel("Number of Attacks")
plt.title("Weapon Type Usage")
plt.show()
```

```python
death_year = df.groupby("iyear")["nkill"].sum()

plt.figure(figsize=(8,5))
plt.plot(death_year.index, death_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.title("Year-wise Deaths Due to Terrorism")
plt.show()
```

```python
region_countries = df.groupby("region_txt")["country_txt"].nunique()

plt.figure(figsize=(10,6))
plt.bar(region_countries.index, region_countries.values)
plt.xlabel("Region")
plt.ylabel("Number of Countries")
plt.title("Region-wise Countries Affected by Terrorism")
plt.xticks(rotation=90)
plt.show()
```
